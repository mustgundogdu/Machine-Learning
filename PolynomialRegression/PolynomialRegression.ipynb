```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

class PolynomialReg:
        
    def __init__(self):
        
        self.linear = self.LinearData()
        
    def LoadCsvData(self):
        datacsv = pd.read_csv('newsalary.csv')
        # Independent variable (score)
        Score = datacsv[['Puan']]
        #dependent varaible (salary)
        Salary = datacsv[['maas']]
        self.linear.DataSplit(Score, Salary)
        
    class LinearData:
        def DataSplit(self, Score, Salary):
            from sklearn.model_selection import train_test_split
            x_train, x_test, y_train, y_test = train_test_split(Score, Salary, test_size=0.33, random_state=0) 
            linearReg = LinearRegression()
            self.PolynomialReg(x_train, x_test, y_train, y_test, linearReg)
            
            
        def PolynomialReg(self, x_train, x_test, y_train, y_test, linearReg):
            polynomialReg = PolynomialFeatures(degree = 6)
            xpolynomial = polynomialReg.fit_transform(x_train)
            
            # Linear Regression object for poly Visulazition
            linearReg.fit(xpolynomial, y_train)
            plt.title("Polynomial Regression Example with Salary DataFrame")
            plt.xlabel("S")
            plt.ylabel("Salary")
            plt.scatter(x_train, y_train, color = 'red')
            plt.plot(x_train, linearReg.predict(polynomialReg.fit_transform(x_train)), color='blue')
            plt.show()
        
    
polobject = PolynomialReg()
polobject.LoadCsvData()
```
