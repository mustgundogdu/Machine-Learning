```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

class RandomForest:
    
    def __init__(self):
        
        self.rForest = self.RandomForestRegressor()
    def LoadDataCsv(self):
        #
        datacsv = pd.read_csv('newsalary.csv')
        Score = datacsv[['Puan']]
        Salary = datacsv[['maas']]
        X = Score.values
        y = Salary.values
        self.rForest.DataSplit(X, y)
    
            
    class RandomForestRegressor:
                     
        def DataSplit(self, X, y):   
            from sklearn.model_selection import train_test_split
            x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
            self.ForestProcess(x_train, y_train)
            
        def ForestProcess(self, X, Y):
            from sklearn.ensemble import RandomForestRegressor
            randomReg = RandomForestRegressor(n_estimators = 10, random_state=0)
            Z = X + 0.5
            K = X - 0.4
            
            randomReg.fit(X, Y.ravel())
            plt.scatter(X,Y, color = 'red')
            plt.plot(X, randomReg.predict(X), color = 'blue')
            plt.plot(X,randomReg.predict(Z), color='green')
            plt.plot(X, randomReg.predict(K), color='yellow')
            
            
svrO = RandomForest()
svrO.LoadDataCsv()
```
